Multivariate Analysis

```python
# Salary and year
years = df['work_year'].unique()  # Get unique years
data = [df[df['work_year'] == year]['salary'] for year in years]  # Group salaries by year

# Plotting the boxplot
plt.figure(figsize=(10, 6))
plt.boxplot(data, labels=years, patch_artist=True, boxprops=dict(facecolor='lightblue'))
plt.title("Salary Distribution by Year", fontsize=15, color='darkblue')
plt.xlabel("Work Year", fontsize=12, color='darkgreen')
plt.ylabel("Salary", fontsize=12, color='darkgreen')
plt.xticks(rotation=45, fontsize=10)
plt.show()
```

```python
# Salary and experience level
experience_levels = df['experience_level'].unique()  # Get unique experience levels
data = [df[df['experience_level'] == level]['salary'] for level in experience_levels]  # Group salaries by experience level

plt.figure(figsize=(8, 5))
plt.boxplot(data, labels=experience_levels, patch_artist=True, boxprops=dict(facecolor='lightgreen'))
plt.title("Salary Distribution by Experience Level", fontsize=15, color='darkblue')
plt.xlabel("Experience Level", fontsize=12, color='darkgreen')
plt.ylabel("Salary", fontsize=12, color='darkgreen')
plt.xticks(rotation=45, fontsize=10)
plt.tight_layout()
plt.show()
```

```python
# Highest paid roles

highest_paid_roles = df.groupby('job_title')['salary'].max().sort_values(ascending=False).head(10)
plt.figure(figsize=(10, 6))
highest_paid_roles.plot(kind='bar', color='goldenrod')
plt.title('Top 10 Highest Paid Roles', fontsize=15, color='darkblue')
plt.xlabel('Job Title', fontsize=12, color='darkgreen')
plt.ylabel('Average Salary', fontsize=12, color='darkgreen')
plt.xticks(rotation=45, fontsize=10, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Experience level by company size

# Preparing the data: Count the number of occurrences for each combination of experience level and company size
experience_by_company_size = df.groupby(['company_size', 'experience_level']).size().unstack()

# Plotting the grouped bar chart
experience_by_company_size.plot(kind='bar', figsize=(10, 6), color=['lightblue', 'orange', 'green', 'red'])
plt.title('Experience Level by Company Size', fontsize=15, color='darkblue')
plt.xlabel('Company Size', fontsize=12, color='darkgreen')
plt.ylabel('Count', fontsize=12, color='darkgreen')
plt.xticks(rotation=0, fontsize=10)
plt.legend(title='Experience Level', fontsize=10)
plt.tight_layout()
plt.show()
```
